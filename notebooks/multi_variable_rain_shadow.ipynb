# Rain-Shadow Effect Quantification Tool

## A Multi-Variable Analysis Framework for Mountain Regions

## Overview

This tool provides a comprehensive framework for quantifying the rain-shadow effect across multiple meteorological variables. It goes beyond traditional precipitation-only analyses to examine how mountain barriers influence temperature, solar radiation, snow depth, and wind patterns on their windward and leeward sides.

The framework was developed and tested for the Wasatch Front and Uinta Basin regions in Utah, but can be adapted for any mountain range where sufficient meteorological data is available.

## Features

- Multi-variable analysis (precipitation, snow depth, air temperature, solar radiation, wind speed)
- Flow-direction classification (westerly, easterly, other)
- Automated data retrieval from weather station networks via Synoptic API
- Data chunking for handling large time periods efficiently
- Robust error handling for inconsistent data structures
- Comprehensive visualization of results
- Automated interpretation of rain-shadow effects for each variable

## Methodology

The analysis follows a structured approach:

1. **Flow Classification**: Wind direction data is retrieved and classified into westerly (225¬∞-315¬∞), easterly (45¬∞-135¬∞), or other flow types
2. **Multi-Variable Retrieval**: Data for precipitation, snow depth, temperature, solar radiation, and wind speed is retrieved for stations on both sides of the mountain
3. **Regional Aggregation**: Station data is grouped by region to create daily averages
4. **Flow-Type Analysis**: Statistics are calculated for each variable by flow type
5. **Interpretation**: Results are analyzed to determine if patterns conform to expected rain-shadow effects

## Output

The tool generates:

1. **Visualizations**: Bar charts comparing each variable by region and flow type
2. **Summary Statistics**: Tabular output of average values and ratios
3. **Interpretation**: Automated analysis of whether patterns conform to expected rain-shadow effects

## Example Results

The analysis quantifies several key aspects of the rain-shadow effect:

- **Precipitation Differential**: During westerly flow, the Wasatch Front receives 2.64x more precipitation than the Uinta Basin, confirming a classic rain-shadow effect
- **Solar Radiation Enhancement**: The Uinta Basin receives 1.38x more solar radiation during westerly flow, consistent with clearer skies in a rain shadow
- **Temperature Patterns**: Contrary to expectations, the Wasatch Front maintains higher temperatures regardless of flow direction
- **Complex Snow Distribution**: Despite less precipitation, the Uinta Basin shows more snow depth during westerly flow
- **Wind Speed Anomalies**: The Uinta Basin consistently experiences stronger winds, contradicting expected sheltering effects

````python
import synoptic
import polars as pl
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import pandas as pd
from tqdm import tqdm

def fetch_variable_data_in_chunks(start_date, end_date, stations_dict, variable, chunk_size=90):
    """
    Fetch data for a specific variable using Synoptic API with improved handling for inconsistent data
    """
    print(f"Fetching {variable} data in chunks...")

    # Combine station IDs from both regions
    wasatch_stations = stations_dict['wasatch_front_stations'].get(variable, [])
    uinta_stations = stations_dict['uinta_basin_stations'].get(variable, [])
    all_stations = ','.join(wasatch_stations + uinta_stations)

    if not all_stations:
        print(f"No stations defined for {variable}")
        return None

    # For large date ranges, split into smaller chunks
    all_data = []
    current_start = start_date

    # Calculate number of chunks
    total_days = (end_date - start_date).days
    num_chunks = (total_days // chunk_size) + 1

    print(f"Fetching data in {num_chunks} chunks of {chunk_size} days each...")

    # Track all column names across chunks to ensure consistency
    all_columns = set()

    for i in range(num_chunks):
        current_end = min(current_start + timedelta(days=chunk_size), end_date)

        print(f"Fetching chunk {i+1}/{num_chunks}: {current_start.strftime('%Y-%m-%d')} to {current_end.strftime('%Y-%m-%d')}")

        try:
            chunk_df = synoptic.TimeSeries(
                stid=all_stations,
                start=current_start,
                end=current_end,
                vars=variable,
                units="metric",
                token="YOUR_TOKEN_HERE"  # Replace with your Synoptic API token
            ).df()

            if not chunk_df.is_empty():
                # Store column names from this chunk
                all_columns.update(chunk_df.columns)
                all_data.append(chunk_df)
                print(f"  ‚Üí Retrieved data for {len(chunk_df.unique('stid'))} stations")
            else:
                print(f"  ‚Üí No data retrieved for this period")

        except Exception as e:
            print(f"  ‚Üí Error fetching data: {e}")

        # Move to next chunk
        current_start = current_end

    # Combine all chunks
    if not all_data:
        print(f"No {variable} data retrieved")
        return None

    print(f"Combining {len(all_data)} data chunks...")

    # Ensure all chunks have the same column structure
    standardized_chunks = []
    for chunk in all_data:
        # Get missing columns in this chunk
        missing_columns = all_columns - set(chunk.columns)
        if missing_columns:
            # Add missing columns with null values
            for col in missing_columns:
                chunk = chunk.with_columns(pl.lit(None).alias(col))
        # Select all columns in a consistent order
        chunk = chunk.select(list(all_columns))
        standardized_chunks.append(chunk)

    # Now concatenate the standardized chunks
    df = pl.concat(standardized_chunks)

    # Add region information
    df = df.with_columns(
        pl.when(pl.col('stid').is_in(wasatch_stations))
        .then(pl.lit('Wasatch Front'))
        .otherwise(pl.lit('Uinta Basin'))
        .alias('region')
    )

    return df

def process_variable_data(var_df, variable):
    """
    Process data for a specific variable with improved error handling
    """
    print(f"Processing {variable} data...")

    # Add date column
    var_df = var_df.with_columns(
        pl.col('date_time').dt.date().alias('date')
    )

    # Group by date and region to get daily regional average
    daily_regional_avg = var_df.group_by(['date', 'region']).agg([
        pl.col('value').mean().alias(f'{variable}_value')
    ])

    # Handle potential missing data
    # Check if both regions are present
    regions = daily_regional_avg.select('region').unique()
    region_list = regions.to_dict(as_series=False)['region']

    if len(region_list) < 2:
        print(f"Warning: Only found data for regions: {region_list}")
        # Add missing region with null values if needed
        all_regions = ['Wasatch Front', 'Uinta Basin']
        missing_regions = [r for r in all_regions if r not in region_list]

        if missing_regions:
            # Get unique dates
            dates = daily_regional_avg.select('date').unique()
            # Create placeholder data for missing regions
            for region in missing_regions:
                placeholder = dates.with_columns([
                    pl.lit(region).alias('region'),
                    pl.lit(0.0).alias(f'{variable}_value')
                ])
                daily_regional_avg = pl.concat([daily_regional_avg, placeholder])

    # Reshape data to have regions as columns
    try:
        pivoted = daily_regional_avg.pivot(
            index='date',
            columns='region',
            values=f'{variable}_value'
        )

        # Rename columns
        if 'Wasatch Front' in pivoted.columns:
            pivoted = pivoted.rename({'Wasatch Front': f'Wasatch_Front_{variable}'})
        else:
            pivoted = pivoted.with_columns(pl.lit(0.0).alias(f'Wasatch_Front_{variable}'))

        if 'Uinta Basin' in pivoted.columns:
            pivoted = pivoted.rename({'Uinta Basin': f'Uinta_Basin_{variable}'})
        else:
            pivoted = pivoted.with_columns(pl.lit(0.0).alias(f'Uinta_Basin_{variable}'))

        # Fill nulls with zeros
        pivoted = pivoted.fill_null(0)
        return pivoted

    except Exception as e:
        print(f"Error pivoting data: {e}")
        # Create an empty DataFrame with the expected structure
        dates = daily_regional_avg.select('date').unique()
        empty_df = dates.with_columns([
            pl.lit(0.0).alias(f'Wasatch_Front_{variable}'),
            pl.lit(0.0).alias(f'Uinta_Basin_{variable}')
        ])
        return empty_df

def analyze_variable_by_flow(regional_var_data, wind_data, variable):
    """
    Analyze variable data by flow type
    """
    print(f"Analyzing {variable} by flow type...")

    # Merge variable data with wind data
    merged_df = regional_var_data.join(
        wind_data.select(['date', 'flow_type']),
        on='date',
        how='inner'
    )

    # Calculate average values by flow type
    flow_avg = merged_df.group_by('flow_type').agg([
        pl.col(f'Wasatch_Front_{variable}').mean().alias('Wasatch_Front_avg'),
        pl.col(f'Uinta_Basin_{variable}').mean().alias('Uinta_Basin_avg')
    ])

    # Calculate ratios
    flow_avg_pd = flow_avg.to_pandas()

    westerly_ratio = None
    easterly_ratio = None

    if 'westerly' in flow_avg_pd['flow_type'].values:
        westerly_row = flow_avg_pd[flow_avg_pd['flow_type'] == 'westerly']
        if westerly_row['Uinta_Basin_avg'].values[0] != 0:
            westerly_ratio = westerly_row['Wasatch_Front_avg'].values[0] / westerly_row['Uinta_Basin_avg'].values[0]

    if 'easterly' in flow_avg_pd['flow_type'].values:
        easterly_row = flow_avg_pd[flow_avg_pd['flow_type'] == 'easterly']
        if easterly_row['Wasatch_Front_avg'].values[0] != 0:
            easterly_ratio = easterly_row['Uinta_Basin_avg'].values[0] / easterly_row['Wasatch_Front_avg'].values[0]

    return {
        'flow_averages': flow_avg,
        'westerly_ratio': westerly_ratio,
        'easterly_ratio': easterly_ratio
    }

def plot_variable_by_flow(variable_results, variable, unit=""):
    """
    Plot variable by flow type
    """
    plt.figure(figsize=(10, 6))
    flow_avg = variable_results['flow_averages'].to_pandas()

    # Set up the bar positions
    bar_width = 0.35
    r1 = np.arange(len(flow_avg))
    r2 = [x + bar_width for x in r1]

    # Create bars
    plt.bar(r1, flow_avg['Wasatch_Front_avg'], width=bar_width, label='Wasatch Front', color='cornflowerblue')
    plt.bar(r2, flow_avg['Uinta_Basin_avg'], width=bar_width, label='Uinta Basin', color='indianred')

    # Add labels and legend
    plt.xlabel('Flow Type')
    plt.ylabel(f'Average {variable.replace("_", " ").title()} {unit}')
    plt.title(f'Average {variable.replace("_", " ").title()} by Region and Flow Type')
    plt.xticks([r + bar_width/2 for r in range(len(flow_avg))], flow_avg['flow_type'])
    plt.legend()

    plt.tight_layout()
    plt.savefig(f'avg_{variable}_by_flow.png', dpi=300)

def analyze_potential_rain_shadow_effects(variable_results, wind_data, variable):
    """
    Analyze if the variable shows a rain-shadow-like effect
    """
    westerly_ratio = variable_results.get('westerly_ratio')
    easterly_ratio = variable_results.get('easterly_ratio')

    effect_description = f"\nAnalysis of {variable.replace('_', ' ').title()}:\n"

    if westerly_ratio is not None:
        if variable in ['air_temp', 'solar_radiation']:
            # For variables where higher values in lee = potential rain shadow effect
            if westerly_ratio < 1:
                effect_description += f"During westerly flow, the Uinta Basin shows {1/westerly_ratio:.2f}x higher values than the Wasatch Front, "
                effect_description += f"which is consistent with a rain-shadow effect (clearer skies, more radiation).\n"
            else:
                effect_description += f"During westerly flow, the Wasatch Front shows {westerly_ratio:.2f}x higher values than the Uinta Basin, "
                effect_description += f"which is contrary to expected rain-shadow patterns.\n"
        else:
            # For variables where lower values in lee = potential rain shadow effect
            if westerly_ratio > 1:
                effect_description += f"During westerly flow, the Wasatch Front shows {westerly_ratio:.2f}x higher values than the Uinta Basin, "
                effect_description += f"which is consistent with a rain-shadow effect.\n"
            else:
                effect_description += f"During westerly flow, the Uinta Basin shows {1/westerly_ratio:.2f}x higher values than the Wasatch Front, "
                effect_description += f"which is contrary to expected rain-shadow patterns.\n"

    if easterly_ratio is not None:
        if variable in ['air_temp', 'solar_radiation']:
            # For variables where higher values in lee = potential rain shadow effect
            if easterly_ratio > 1:
                effect_description += f"During easterly flow, the Uinta Basin shows {easterly_ratio:.2f}x higher values than the Wasatch Front, "
                effect_description += f"which is consistent with a reverse rain-shadow effect.\n"
            else:
                effect_description += f"During easterly flow, the Wasatch Front shows {1/easterly_ratio:.2f}x higher values than the Uinta Basin, "
                effect_description += f"which is contrary to expected reverse rain-shadow patterns.\n"
        else:
            # For variables where lower values in lee = potential rain shadow effect
            if easterly_ratio < 1:
                effect_description += f"During easterly flow, the Uinta Basin shows {easterly_ratio:.2f}x lower values than the Wasatch Front, "
                effect_description += f"which is consistent with a reverse rain-shadow effect.\n"
            else:
                effect_description += f"During easterly flow, the Uinta Basin shows {easterly_ratio:.2f}x higher values than the Wasatch Front, "
                effect_description += f"which is contrary to expected reverse rain-shadow patterns.\n"

    return effect_description

def main():
    """
    Main function to execute the expanded rain-shadow analysis with improved error handling
    """
    print("Expanded Rain-Shadow Effect Analysis for Wasatch Mountains and Uinta Basin")
    print("="*85)

    # Define expanded station dictionaries
    stations_dict = {
        'wasatch_front_stations': {
            'snow_depth': ["KHIF", "KSLC", "UTORM", "COOPPROU1"],
            'air_temp': ["QHW", "QLN", "QHV", "QBV", "QCV"],
            'solar_radiation': ["QHW", "QLN", "QBV"],
            'wind_speed': ["QHW", "QLN", "QHV", "QBV", "QCV"],
            'precip_accum_24_hour': ['KOGD', 'AP611', 'C8948', 'KSLC', 'COOPPROU1'],
            'sea_level_pressure': ['KSLC']
        },
        'uinta_basin_stations': {
            'snow_depth': ["COOPDSNU1", "COOPFTDU1", "COOPJENU1", "COOPVELU1"],
            'air_temp': ["QV4", "QRS", "A1388", "UCL13"],
            'solar_radiation': ["QV4", "UCL13"],
            'wind_speed': ["QV4", "QRS", "UCL13"],
            'precip_accum_24_hour': ['COOPDSNU1', 'COOPMYTU1', 'COOPFTDU1', 'KVEL', 'COOPJENU1'],
            'sea_level_pressure': ['KVEL']
        }
    }

    # Define time period for analysis (e.g., 1 year to keep processing time reasonable)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)

    print(f"Analyzing period: {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}")

    # Store results for each variable
    variable_results = {}

    try:
        # First fetch wind direction data to classify days
        wind_data = fetch_wind_direction_data_in_chunks(start_date, end_date, stations_dict)

        # List of variables to analyze
        variables = ['precip_accum_24_hour', 'snow_depth', 'air_temp', 'solar_radiation', 'wind_speed']

        # Analyze each variable
        for variable in variables:
            try:
                # Fetch and process variable data
                var_data = fetch_variable_data_in_chunks(start_date, end_date, stations_dict, variable)

                if var_data is not None:
                    regional_var_data = process_variable_data(var_data, variable)

                    # Analyze variable by flow type
                    var_results = analyze_variable_by_flow(regional_var_data, wind_data, variable)
                    variable_results[variable] = var_results

                    # Plot results
                    units = {
                        'precip_accum_24_hour': '(mm)',
                        'snow_depth': '(cm)',
                        'air_temp': '(¬∞C)',
                        'solar_radiation': '(W/m¬≤)',
                        'wind_speed': '(m/s)'
                    }
                    plot_variable_by_flow(var_results, variable, units.get(variable, ""))
            except Exception as e:
                print(f"Error processing {variable}: {e}")
                import traceback
                traceback.print_exc()

        # Print findings for each variable
        print("\n" + "="*50)
        print("EXPANDED RAIN-SHADOW EFFECT ANALYSIS RESULTS")
        print("="*50)

        for variable, results in variable_results.items():
            print(f"\n{variable.replace('_', ' ').title()} Analysis")
            print("-"*40)
            print(results['flow_averages'])

            if results['westerly_ratio'] is not None:
                print(f"\nFor westerly flow, the ratio between Wasatch Front and Uinta Basin is {results['westerly_ratio']:.2f}")

            if results['easterly_ratio'] is not None:
                print(f"For easterly flow, the ratio between Uinta Basin and Wasatch Front is {results['easterly_ratio']:.2f}")

        # Analyze potential rain-shadow effects for each variable
        print("\n" + "="*50)
        print("POTENTIAL RAIN-SHADOW EFFECTS BY VARIABLE")
        print("="*50)

        for variable, results in variable_results.items():
            effect_description = analyze_potential_rain_shadow_effects(results, wind_data, variable)
            print(effect_description)

        print("\nAnalysis complete. Visualization files have been saved.")

    except Exception as e:
        print(f"Error during analysis: {e}")
        import traceback
        traceback.print_exc()

# Copy from original code - needed for the wind direction analysis
def fetch_wind_direction_data_in_chunks(start_date, end_date, stations_dict, chunk_size=90):
    """
    Fetch wind direction data using Synoptic API, handling large date ranges
    """
    print("Fetching wind direction data in chunks...")

    # Use Salt Lake City for wind direction
    slc_station = stations_dict['wasatch_front_stations']['sea_level_pressure'][0]

    # For large date ranges, split into smaller chunks
    all_wind_data = []
    current_start = start_date

    # Create progress bar for data fetching
    total_days = (end_date - start_date).days
    num_chunks = (total_days // chunk_size) + 1

    print(f"Fetching wind data in {num_chunks} chunks of {chunk_size} days each...")

    for i in range(num_chunks):
        current_end = min(current_start + timedelta(days=chunk_size), end_date)

        print(f"Fetching chunk {i+1}/{num_chunks}: {current_start.strftime('%Y-%m-%d')} to {current_end.strftime('%Y-%m-%d')}")

        try:
            chunk_df = synoptic.TimeSeries(
                stid=slc_station,
                start=current_start,
                end=current_end,
                vars="wind_direction,wind_speed",
                units="metric",
                token="YOUR_TOKEN_HERE" # Replace with your Synoptic API token
            ).df()

            if not chunk_df.is_empty():
                all_wind_data.append(chunk_df)
                print(f"  ‚Üí Retrieved data for {chunk_df.shape[0]} observations")
            else:
                print(f"  ‚Üí No data retrieved for this period")

        except Exception as e:
            print(f"  ‚Üí Error fetching wind data: {e}")

        # Move to next chunk
        current_start = current_end

    # Combine all chunks
    if not all_wind_data:
        print("No wind data retrieved, creating synthetic data")
        return create_synthetic_wind_data(start_date, end_date)

    print(f"Combining {len(all_wind_data)} wind data chunks...")
    df = pl.concat(all_wind_data)

    # Process wind direction data
    return process_wind_direction_data(df, start_date, end_date)

def process_wind_direction_data(df, start_date, end_date):
    """
    Process wind direction data and classify flow types
    """
    print("Processing wind direction data...")

    # Check if wind_direction is in the data
    if 'wind_direction' not in df.select('variable').unique().to_dict(as_series=False)['variable']:
        print("No wind_direction variable found, using synthetic data")
        return create_synthetic_wind_data(start_date, end_date)

    # Filter for wind direction data only
    wind_dir_df = df.filter(pl.col('variable') == 'wind_direction')

    # Convert to pandas for simpler processing
    wind_dir_df_pd = wind_dir_df.to_pandas()

    # Add flow type classification
    wind_dir_df_pd['flow_type'] = pd.NA  # Initialize with NA
    mask_westerly = (wind_dir_df_pd['value'] >= 225) & (wind_dir_df_pd['value'] <= 315)
    mask_easterly = (wind_dir_df_pd['value'] >= 45) & (wind_dir_df_pd['value'] <= 135)

    wind_dir_df_pd.loc[mask_westerly, 'flow_type'] = 'westerly'
    wind_dir_df_pd.loc[mask_easterly, 'flow_type'] = 'easterly'
    wind_dir_df_pd.loc[~(mask_westerly | mask_easterly), 'flow_type'] = 'other'

    # Add date column
    wind_dir_df_pd['date'] = pd.to_datetime(wind_dir_df_pd['date_time']).dt.date

    # Group by date to get daily flow type
    print("Aggregating wind data by day...")
    daily_wind_pd = wind_dir_df_pd.groupby('date').agg({
        'value': 'mean',
        'flow_type': lambda x: x.mode()[0] if not x.empty and not x.isna().all() else 'other'
    }).reset_index()

    daily_wind_pd.columns = ['date', 'wind_direction_mean', 'flow_type']

    # Check for missing dates
    all_dates = pd.date_range(start_date, end_date).date
    missing_dates = set(all_dates) - set(daily_wind_pd['date'])

    if missing_dates:
        print(f"Found {len(missing_dates)} days with missing wind data. Filling with 'other' flow type.")
        missing_df = pd.DataFrame({
            'date': list(missing_dates),
            'wind_direction_mean': np.nan,
            'flow_type': 'other'
        })
        daily_wind_pd = pd.concat([daily_wind_pd, missing_df]).sort_values('date')

    # Convert back to polars
    daily_wind = pl.from_pandas(daily_wind_pd)

    return daily_wind

def create_synthetic_wind_data(start_date, end_date):
    """
    Create synthetic wind direction data if API fails
    """
    print("Creating synthetic wind direction data")

    # Generate dates
    dates = [start_date.date() + timedelta(days=i) for i in range((end_date - start_date).days + 1)]

    # Create synthetic data with predominantly westerly flow (realistic for the region)
    np.random.seed(42)  # For reproducibility
    flow_types = np.random.choice(['westerly', 'easterly', 'other'], size=len(dates), p=[0.6, 0.1, 0.3])
    wind_dirs = []

    for flow in flow_types:
        if flow == 'westerly':
            wind_dirs.append(np.random.uniform(225, 315))
        elif flow == 'easterly':
            wind_dirs.append(np.random.uniform(45, 135))
        else:
            if np.random.random() < 0.5:
                wind_dirs.append(np.random.uniform(0, 45))
            else:
                wind_dirs.append(np.random.uniform(315, 360))

    # Create fallback DataFrame
    fallback_data = {
        'date': dates,
        'wind_direction_mean': wind_dirs,
        'flow_type': flow_types
    }

    fallback_df = pl.DataFrame(fallback_data)
    print("WARNING: Using synthetic wind directions due to data retrieval error")

    return fallback_df

if __name__ == "__main__":
    main()
````
